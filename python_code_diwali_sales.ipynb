```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
sns.countplot(x = 'Gender', data = df)
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
amount_gender = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

ab = sns.barplot(x = 'Gender', y = 'Amount', data = amount_gender)

for bars in ab.containers:
    ab.bar_label(bars)
```

*From the above graph we can say that most of the buyers are females and even purchasing power of female is much hugher than males*

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

ab = sns.barplot(x = 'Age Group', y = 'Amount', data = amount_age)

for bars in ab.containers:
    ab.bar_label(bars)
```

*From the above graph we can say that most of the buyers are in the age group of 26 to 35 years female*

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(15, 15)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

*From the above graph we can say that mst of the orders are come from Uttar Pradesh, Maharastra and resectively*

```python
state_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(15, 15)})
sns.barplot(data = state_amount, x = 'State', y = 'Amount')
```

*From the above graph we can say that most of the amount of sales are from Uttar Pradesh then Maharastra*

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize':(5, 2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(4, 3)})
sns.barplot(data = sales_marital, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

*From the above graph we can say that most of the buyers are married women and they have high purchasing power*

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(20, 5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

*From the above graph we can see that from the IT sectors workers, healthcare workers had given most of the orders*

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize':(20, 3)})
sns.barplot(data = sales_occupation, x = 'Occupation', y = 'Amount')
```

*From the above graph we can say that most of the buyers are from IT sectors, healthcare systems, and aviation.*

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(25, 5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see that the most selling category are clothing and then food category

```python
sales_product = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize':(25, 3)})
sns.barplot(data = sales_product, x = 'Product_Category', y = 'Amount')
```

From the above graph we can say that most of the selling products are from food then clothing catagories

```python
df.columns
```

```python
product_id = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(15)

sns.set(rc = {'figure.figsize':(25, 3)})
sns.barplot(data = product_id, x = 'Product_ID', y = 'Orders')
```
